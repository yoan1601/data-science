Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...
```

1. Finding schools with the best math scores (average_math >= 80% of 800)

```python
best_math_schools_full = schools[schools["average_math"] >= ((80 * 800 )/ 100)].sort_values("average_math", ascending = False)
# print(best_math_schools.loc[:, "school_name":"average_math"])
top_10_schools_full = schools[["school_name", "borough"]]
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
top_10_schools_full["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
top_10_schools_full = top_10_schools_full.sort_values("total_SAT", ascending = False)
top_10_schools_full = top_10_schools_full.iloc[:10]
```

```python
best_math_schools = best_math_schools_full[["school_name", "average_math"]]
best_math_schools
```

```python
top_10_schools_full[["total_SAT"]].fillna(0) 
top_10_schools_full
```

```python
top_10_schools = top_10_schools_full[["school_name", "total_SAT"]]
top_10_schools
```

```python
borough_std = top_10_schools_full.groupby("borough")[["total_SAT"]].std()
borough_std = borough_std.sort_values("total_SAT", ascending = False)
print(borough_std)
top_borough = borough_std.reset_index()
top_borough = top_borough.iloc[0,0]
top_borough
```

```python
all_schools_in_top_borough = schools[schools["borough"] == top_borough]
all_schools_in_top_borough
```

```python
top_schools_in_top_borough = top_10_schools_full[top_10_schools_full["borough"] == top_borough]
```

```python
top_num_schools = len(all_schools_in_top_borough)
top_num_schools
```

```python
top_average_SAT = all_schools_in_top_borough["total_SAT"].mean()
top_average_SAT
```

```python
top_std_SAT = all_schools_in_top_borough["total_SAT"].std()
top_std_SAT
```

```python
largest_std_dev = {
    "borough" : [top_borough],
    "num_schools" : [top_num_schools],
    "average_SAT" : [round(top_average_SAT, 2)],
    "std_SAT" : [round(top_std_SAT, 2)]
}

largest_std_dev = pd.DataFrame(largest_std_dev)
largest_std_dev
```
