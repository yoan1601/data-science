Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# Import necessary packages
import pandas as pd
import numpy as np
```

**1 - Loading the data**

```python
airbnb_price = pd.read_csv('data/airbnb_price.csv')
airbnb_price.head()
```

```python
airbnb_room_type = pd.read_excel('data/airbnb_room_type.xlsx')
airbnb_room_type.head()
```

```python
airbnb_last_review = pd.read_csv('data/airbnb_last_review.tsv', sep = '\t')
airbnb_last_review.head()
```

**2 - Merging the three DataFrames**

```python
airbnb = airbnb_price.merge(airbnb_room_type, on = 'listing_id', how = 'inner')
airbnb = airbnb.merge(airbnb_last_review, on = 'listing_id', how = 'inner')
airbnb.head()
```

**3 - Determining the earliest and most recent review dates**

**3-a - Converting reviews data to a date format**

```python
airbnb['last_review_date'] = pd.to_datetime(airbnb['last_review'])
airbnb['last_review_date'].head()
```

**3-b - Finding the earliest and most recent dates**

```python
earliest_review_date = airbnb['last_review_date'].min().date()
most_recent_review_date = airbnb['last_review_date'].max().date()
print([earliest_review_date, most_recent_review_date])
```

**4 - Finding how many listings are private rooms**

```python
airbnb['room_type'].unique()
```

**4-a - Cleaning data with pandas**

```python
airbnb['room_type'] = airbnb['room_type'].str.lower()
airbnb['room_type'].unique()
```

**4-b - Counting the number of private rooms**

```python
private_rooms_count = len(airbnb[airbnb['room_type'] == 'private room'])
private_rooms_count
```

**5 - Finding the average price of listings**

**5-a - Converting to numeric type**

```python
airbnb['price'] = airbnb['price'].str.replace(" dollars", "").astype("float")
```

```python
airbnb.info()
```

**5-b - Finding the mean**

```python
avg_price = round(airbnb['price'].mean(), 2)
avg_price
```

**6 - Creating a DataFrame with the four solution values**

```python
data = {
    "first_reviewed": [earliest_review_date],
    "last_reviewed": [most_recent_review_date],
    "nb_private_rooms": [private_rooms_count],
    "avg_price": [avg_price]
}
review_dates = pd.DataFrame(data)
review_dates
```
